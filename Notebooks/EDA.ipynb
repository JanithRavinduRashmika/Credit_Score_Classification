```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split

unique_customer_ids = df['Customer_ID'].unique()

train_customer_ids, test_customer_ids = train_test_split(unique_customer_ids, test_size=0.2, random_state=42)

train_df = df[df['Customer_ID'].isin(train_customer_ids)]
test_df = df[df['Customer_ID'].isin(test_customer_ids)]

print("Train DataFrame:")
print(train_df)
print("\nTest DataFrame:")
print(test_df)
```

#### Correlation

```python
from scipy.stats import spearmanr

# Initialize a dictionary to store Spearman's correlation coefficients
spearman_corr = {}

# Compute Spearman's Rank Correlation for each numerical variable with the target variable
for col in train_df.select_dtypes(include=['float64', 'int64']).columns:
    if col != 'Credit_Score':  # Exclude the target variable itself
        corr, _ = spearmanr(train_df[col], train_df['Credit_Score'])
        spearman_corr[col] = corr

# Print the Spearman's Rank Correlation coefficients
print("Spearman's Rank Correlation coefficients:")
for var, corr in spearman_corr.items():
    print(f"{var}: {corr:.4f}")
```

```python
correlation_matrix = train_df[numeric_col].corr(method='pearson')

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Oranges', fmt='.2f')
plt.title('Pearson Correlation Matrix')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cross_decomposition import PLSRegression

target_mapping = {'Poor':0, 'Standard':1, 'Good':2}

# Sample data (replace this with your own dataset)
X = train_df[numeric_col]
y = train_df['Credit_Score'].map(target_mapping)

# Create and fit the PLS model
model1 = PLSRegression(n_components=2)
model1.fit(X, y)

# Summary of the PLS model (equivalent to summary(model1) in R)
print("PLS Summary:")
print("R-squared (calibration):", model1.score(X, y))
#print("R-squared (test):", model1.score(xt, yt))
```

```python
plt.figure(figsize=(10, 8))

# List of unique classes/categories in `y`
classes = np.unique(y)

# Colors for each class
class_colors = {
    0: '#ffff00',
    1: '#000000',
    2: '#f04a00'
}

# Mapping from numerical to descriptive labels
class_descriptions = {
    0: 'Poor',
    1: 'Standard',
    2: 'Good'
}

# For each class in the descriptions (ensures all classes are handled)
for class_num, description in class_descriptions.items():
    # If the current class is not in y_train, continue without plotting
    if class_num not in classes:
        continue

    # Get the scores for the samples that belong to this class/category
    x_scores_class = model1.x_scores_[y == class_num, 0]
    y_scores_class = model1.x_scores_[y == class_num, 1]

    # Scatter plot for this class/category using the descriptive label
    plt.scatter(x_scores_class, y_scores_class, label=description, color=class_colors[class_num])

plt.xlabel('Comp 1 (Score 1)')
plt.ylabel('Comp 2 (Score 2)')
plt.title('PLS Scores Plot')
plt.legend()  # Add a legend
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.scatter(model1.x_loadings_[:, 0], model1.x_loadings_[:, 1], s=200)
plt.xlabel('Loading 1')
plt.ylabel('Loading 2')
plt.title('PLS Loadings Plot')
plt.rcParams['font.size'] = 10    # Set the default font size for all elements

# Annotate the data points with feature names
for i, feature_name in enumerate(X.columns):
    plt.annotate(feature_name, (model1.x_loadings_[i, 0], model1.x_loadings_[i, 1]))


plt.scatter(model1.y_loadings_[:, 0], model1.y_loadings_[:, 1], s = 200)
plt.annotate('Credit_Score', (model1.y_loadings_[:, 0], model1.y_loadings_[:, 1]))

plt.grid(True)
plt.show()
```
