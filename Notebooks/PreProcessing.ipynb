```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Data/cleaned_train.csv")
```

```python
# drop ID, Customer_ID, Month, Name, SSN

df = df.drop(["ID", "Customer_ID", "Month", "Name", "SSN", "Occupation"], axis=1)
```

```python
df = df.drop_duplicates()
df.shape

# there are no duplicates
```

```python
df.T
```

```python
df_notEncoded = df.copy()
df_notEncoded["Spending_Behaviour"] = df_notEncoded["Spending_Behaviour"].map({1:"High", 0:"Low"})
df_notEncoded["Paying_Behaviour"] = df_notEncoded["Paying_Behaviour"].map({2:"Large_value_payments", 1:"Medium_value_payments", 0:"Small_value_payments"})
df_notEncoded["Payment_of_Min_Amount"] = df_notEncoded["Payment_of_Min_Amount"].map({0:"No", 1:"Yes"})
df_notEncoded["Credit_Mix"] = df_notEncoded["Credit_Mix"].map({2:"Good", 1:"Standard", 0:"Bad"})
```

```python
df["Credit_Score"] = df["Credit_Score"].map({"Good":2, "Standard":1, "Poor":0})
```

```python
X = df.drop("Credit_Score", axis=1)
y = df["Credit_Score"]
```

```python
categorical_features = ["Credit_Mix", "Payment_of_Min_Amount" , "Spending_Behaviour", "Paying_Behaviour"]
numerical_features = ['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
       'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Type_of_Loan',
       'Delay_from_due_date', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit',
       'Num_Credit_Inquiries', 'Outstanding_Debt',
       'Credit_Utilization_Ratio', 'Credit_History_Age',
       'Total_EMI_per_month',
       'Amount_invested_monthly', 'Monthly_Balance']
```

```python
from sklearn.feature_selection import mutual_info_classif

# Compute mutual information
mi = mutual_info_classif(X, y)

# Create a DataFrame to rank features
mi_df = pd.DataFrame({'Feature': X.columns, 'Mutual Information': mi})
mi_df = mi_df.sort_values(by='Mutual Information', ascending=False)

print(mi_df)
```
