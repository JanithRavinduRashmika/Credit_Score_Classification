```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../Data/cleaned_train.csv", low_memory=False)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

unique_customer_ids = df['Customer_ID'].unique()

train_customer_ids, test_customer_ids = train_test_split(unique_customer_ids, test_size=0.2, random_state=42)

train_df = df[df['Customer_ID'].isin(train_customer_ids)]
test_df = df[df['Customer_ID'].isin(test_customer_ids)]

print("Train DataFrame:")
print(train_df)
print("\nTest DataFrame:")
print(test_df)
```

#### Correlation

```python
from scipy.stats import spearmanr

# Initialize a dictionary to store Spearman's correlation coefficients
spearman_corr = {}

# Compute Spearman's Rank Correlation for each numerical variable with the target variable
for col in train_df.select_dtypes(include=['float64', 'int64']).columns:
    if col != 'Credit_Score':  # Exclude the target variable itself
        corr, _ = spearmanr(train_df[col], train_df['Credit_Score'])
        spearman_corr[col] = corr

# Print the Spearman's Rank Correlation coefficients
print("Spearman's Rank Correlation coefficients:")
for var, corr in spearman_corr.items():
    print(f"{var}: {corr:.4f}")
```

```python
correlation_matrix = train_df[numeric_col].corr(method='pearson')

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt='.2f')
plt.title('Pearson Correlation Matrix')
plt.show()
```
